```python
python_pass = ({"sunil", "anil", "Chandu", "Paddu", "sunny"})
sql_pass = ({"bunny", "chinni","sunil", "anil", "chandu"})
print("Union :", python_pass | sql_pass)
print("intersection :", python_pass & sql_pass)
print("Diff :", python_pass - sql_pass)
```

```python
import json

student = {
    "name" : "Sunil Kumar",
    "age" : 48,
    "passed" : True,
    "marks" : [90, 91, 92],
    "address" : {
        "city" : "Kurnool",
        "pincode" : 518004
         }
    }
student_json = json.dumps(student, indent = 4)
print(student_json)
```

```python
value = json.loads(student_json)
print(f"name : {value['name']}")
print(f"Second marks : {value['marks'][1]}")
print(f"city : {value['address']['city']}")
```
